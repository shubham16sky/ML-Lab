K - means and EM algorithm

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = datasets.load_iris()

x = pd.DataFrame(data.data)
y = pd.DataFrame(data.target)
x.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']


model = KMeans(n_clusters=3).fit(x.iloc[:,:-1])
gmm = GaussianMixture(n_components=3).fit(x.iloc[:,:-1])




#real classification plotting
plt.subplot(2,2,1)
plt.title("real")
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.scatter(x.Petal_Length,x.Petal_Width,color="red")

#K means plotting 
model = KMeans(n_clusters=3).fit(x.iloc[:,:-1])
plt.subplot(2,2,2)
plt.title("K-Means")
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.scatter(x.Petal_Length,x.Petal_Width,color="blue")

#Gaussian Mixture classification 
gmm = GaussianMixture(n_components=3).fit(x.iloc[:,:-1])
plt.subplot(2,2,3)
plt.title("Gaussin Mixture")
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.scatter(x.Petal_Length,x.Petal_Width,color="green")

plt.show()
```
